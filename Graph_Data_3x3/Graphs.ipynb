```python
import matplotlib.pyplot as plt
import pickle
```

```python
#DDQN 3x3
x = pickle.load( open( "32DDQN_41472_loss_results.pickle", "rb" ) )
plt.figure
plt.plot(x)
plt.ylabel('Loss')
plt.xlabel("Episode Number")
plt.tight_layout()
plt.savefig("DDQN_Loss_3_3.eps", format = "eps")
plt.show()

y = pickle.load( open( "32DDQN_41472_rewards_results.pickle", "rb" ) )
plt.figure
plt.plot(y)
plt.ylabel('Collected Reward')
plt.xlabel("Episode Number")
plt.tight_layout()
plt.savefig("DDQN_Rewards_3_3.eps", format = "eps")
plt.show()
```
